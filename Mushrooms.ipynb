```python
import pandas as pd
import numpy as np
```

```python
mushroom = pd.read_csv("mushroom.csv")
mushroom.columns.values
```

```python
mushroom['Cap-shape'] = mushroom['Cap-shape'].map({'x':0, 'b':1, 's':2, 'f':3, 'k':4, 'c':5})
mushroom[' Cap-surface'] = mushroom[' Cap-surface'].map({'s':0, 'y':1, 'f':2, 'g':3})
mushroom[' Cap-color'] = mushroom[' Cap-color'].map({'n':0, 'y':1, 'w':2, 'g':3, 'e':4, 'p':5, 'b':6, 'u':7, 'c':8, 'r':9})
mushroom[' Bruises'] = mushroom[' Bruises'].map({'f':0, 't':1})
mushroom[' Odor'] = mushroom[' Odor'].map({'p':0, 'a':1, 'l':2, 'n':3, 'f':4, 'c':5, 'y':6, 's':7, 'm':8})
mushroom[' Gill-attachment'] = mushroom[' Gill-attachment'].map({'f':0, 'a':1, 'd':2, 'n':3})
mushroom[' Gill-size'] = mushroom[' Gill-size'].map({'n':0, 'b':1})
mushroom[' Gill-spacing'] = mushroom[' Gill-spacing'].map({'c':0, 'w':1, 'd':2})
mushroom[' Gill-color'] = mushroom[' Gill-color'].map({'k':0, 'n':1, 'w':2, 'h':3, 'u':4, 'b':5, 'r':6, 'y':7, 'o':8, 'g':9, 'p':10, 'e':11})
mushroom[' Stalk-shape'] = mushroom[' Stalk-shape'].map({'e':0, 't':1})
mushroom[' Stalk-root'] = mushroom[' Stalk-root'].map({'e':0, 'c':1, 'b':2, 'r':3, 'u':4, 'z':5, '?':None})
mushroom[' Stalk-surface-above-ring'] = mushroom[' Stalk-surface-above-ring'].map({'f':0, 'k':1, 's':2, 'y':3})
mushroom[' Stalk-surface-below-ring'] = mushroom[' Stalk-surface-below-ring'].map({'f':0, 'k':1, 's':2, 'y':3})
mushroom[' Stalk-color-above-ring'] = mushroom[' Stalk-color-above-ring'].map({'w':0, 'g':1, 'p':2, 'n':3, 'b':4, 'e':5, 'o':6, 'c':7, 'y':8})
mushroom[' Stalk-color-below-ring'] = mushroom[' Stalk-color-below-ring'].map({'w':0, 'g':1, 'p':2, 'n':3, 'b':4, 'e':5, 'o':6, 'c':7, 'y':8})
mushroom[' Veil-type'] = mushroom[' Veil-type'].map({'u':0, 'p':1})
mushroom[' Veil-color'] = mushroom[' Veil-color'].map({'w':0, 'n':1, 'o':2, 'y':3})
mushroom[' Ring-number'] = mushroom[' Ring-number'].map({'t':0, 'n':1, 'o':2})
mushroom[' Ring-type'] = mushroom[' Ring-type'].map({'p':0, 'e':1, 'l':2, 'f':3, 'n':4, 'c':5, 's':6, 'z':7})
mushroom[' Spore-print-color'] = mushroom[' Spore-print-color'].map({'k':0, 'n':1, 'w':2, 'h':3, 'u':4, 'b':5, 'r':6, 'y':7, 'o':8})
mushroom[' Population'] = mushroom[' Population'].map({'s':0, 'n':1, 'a':2, 'v':3, 'y':4, 'c':5})
mushroom[' Habitat'] = mushroom[' Habitat'].map({'u':0, 'g':1, 'm':2, 'd':3, 'p':4, 'w':5, 'l':6})
```

```python
mushroom
```

```python
mushroom = mushroom.dropna(how='any',axis=0)
mushroom.shape
```

```python
mushroom.to_csv('mushroom_numerical.csv', sep=',', encoding='utf-8', index=False)
```
