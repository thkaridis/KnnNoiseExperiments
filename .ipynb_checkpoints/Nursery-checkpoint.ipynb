```python
import pandas as pd
import numpy as np
```

```python
nursery = pd.read_csv("nursery.csv")
nursery.columns.values
```

```python
for index, row in nursery.iterrows():
    if row[0] == 'usual':
        row[0] = 1
    elif row[0] == 'pretentious':
        row[0] = 2
    elif row[0] == 'great_pret':
        row[0] = 3
    
    if row[1] == ' proper':
        row[1] = 1
    elif row[1] == ' less_proper':
        row[1] = 2
    elif row[1] == ' improper':
        row[1] = 3
    elif row[1] == ' critical':
        row[1] = 4
    elif row[1] == ' very_crit':
        row[1] = 5
    
    if row[2] == ' complete':
        row[2] = 1
    elif row[2] == ' incomplete':
        row[2] = 2
    elif row[2] == ' foster':
        row[2] = 3
    elif row[2] == ' critical':
        row[2] = 4
    elif row[2] == ' completed':
        row[2] = 5
       
    if row[3] == ' more':
        row[3] = 10
        
    if row[4] == ' convenient':
        row[4] = 1
    elif row[4] == ' less_conv':
        row[4] = 2
    elif row[4] == ' critical':
        row[4] = 3
    
    if row[5] == ' convenient':
        row[5] = 1
    elif row[5] == ' inconv':
        row[5] = 2
   
    if row[6] == ' nonprob':
        row[6] = 1
    elif row[6] == ' slightly_prob':
        row[6] = 2
    elif row[6] == ' problematic':
        row[6] = 2
        
    if row[7] == ' recommended':
        row[7] = 1
    elif row[7] == ' not_recom':
        row[7] = 2
    elif row[7] == ' priority':
        row[7] = 2
```
