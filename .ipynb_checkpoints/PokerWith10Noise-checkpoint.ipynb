```python
i = 100
j = 0
features = ['S1', ' C1', ' S2', ' C2', ' S3', ' C3', ' S4', ' C4', ' S5', ' C5']
for index, m in poker.iterrows():
    if index % 20 == 0:
        poker.at[index,features[j]] = i + 1
        j += 3
        i+=10
        if j >= 9:
            j = 0
```

```python
i = 0
j = 0
for index, m in poker.iterrows():
    if index % 20 == 0:
        if index < 1025003:
            poker.at[index+5,:] = poker.loc[j,:]
        else:
            poker.at[index+1,:] = poker.loc[j,:]
        j += 120
        if j >= 1025008:
            j = 0
```
