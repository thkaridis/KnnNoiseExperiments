```python
import pandas as pd
import numpy as np
```

```python
breast = pd.read_csv("breastCancer.csv")
breast.columns.values
#breast = breast.dropna(how='any',axis=0)
```

```python
breast['age'] = breast['age'].map({'10-19':0, '20-29':1, '30-39':2, '40-49':3, '50-59':4, '60-69':5, '70-79':6, '80-89':7, '90-99':8})
breast['menopause'] = breast['menopause'].map({'lt40':0, 'ge40':1, 'premeno':2})
breast['tumor-size'] = breast['tumor-size'].map({'0-4':0, '5-9':1, '10-14':2, '15-19':3, '20-24':4, '25-29':5, '30-34':6, '40-44':7, '45-49':8, '50-54':9, '55-59':10})
breast['inv-nodes'] = breast['inv-nodes'].map({'0-2':0, '3-5':1, '6-8':2, '9-11':3, '12-14':4, '15-17':5, '18-20':6, '21-23':7, '24-26':8, '27-29':9, '30-32':10, '33-35':11, '36-39':12})
breast['node-caps'] = breast['node-caps'].map({'yes':0, 'no':1})
breast['deg-malig'] = breast['deg-malig']
breast['breast'] = breast['breast'].map({'left':0, 'right':1})
breast['breast-quad'] = breast['breast-quad'].map({'left_up':0, 'left_low':1, 'right_up':2, 'right_low':2, 'central':2})
breast['irradiat'] = breast['irradiat'].map({'yes':0, 'no':1})
```

```python
breast
```

```python
breast = breast.dropna(how='any',axis=0)
breast.shape
```

```python
breast['age'] = breast['age'].astype(int)
breast['menopause'] = breast['menopause'].astype(int)
breast['tumor-size'] = breast['tumor-size'].astype(int)
breast['inv-nodes'] = breast['inv-nodes'].astype(int)
breast['node-caps'] = breast['node-caps'].astype(int)
breast['deg-malig'] = breast['deg-malig'].astype(int)
breast['breast'] = breast['breast'].astype(int)
breast['breast-quad'] = breast['breast-quad'].astype(int)
breast['irradiat'] = breast['irradiat'].astype(int)
breast
```

```python
breast.to_csv('breast_numerical.csv', sep=',', encoding='utf-8', index=False)
```
